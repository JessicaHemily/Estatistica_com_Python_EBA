```python
import pandas as pd
import seaborn as sns #manipular gráfico
import numpy as np #operações numpy
import matplotlib.pyplot as plt #manipular gráfico
```

```python
df = pd.read_csv("house_price.csv")
df
```

```python
df.describe()
```

```python
df.info()
```

```python
mean = df["SalePrice"].mean()
print(f"A média de VENDAS é: {mean:.2f}$")
```

```python
median = df["SalePrice"].median()
print(f"A mediana dos valores de VENDAS é: {median:.2f}$")
```

```python
mode = df["SalePrice"].mode()
print(f"A moda dos valores de VENDAS é: {mode}$")
```

```python
print("media:",mean,"mediana:",median,"moda:",mode)
```

```python
sns.boxplot(df.SalePrice)
```

```python
df["categoria_preco"] = np.where(df["SalePrice"]>214000,"cara","barata")
df
```

```python
print("\033[1mMédia da nota geral:")
df.groupby("categoria_preco")["OverallQual"].mean()
```

```python
print("\033[1mMediana da nota geral:")
df.groupby("categoria_preco")["OverallQual"].median()
```

```python
sns.boxplot( x="categoria_preco", y="OverallQual", data=df)
```
