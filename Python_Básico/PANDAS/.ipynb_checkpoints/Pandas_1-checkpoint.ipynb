```python
df2.groupby('artist')['title'].count()
```

```python
df_teste = df2.groupby('artist')['title'].count()
df_teste = df_teste.reset_index()
df_teste = df_teste.rename(columns = {'title':'contagem_de_title'})
df_teste = df_teste.sort_values(by='contagem_de_title', ascending = False)
df_teste
```

```python
df_contagem_musicas = df2.groupby(['artist','year'])['title'].count()
df_contagem_musicas = df_contagem_musicas.reset_index()
df_contagem_musicas = df_contagem_musicas.rename(columns = {'title':'contagem_title'})
df_contagem_musicas = df_contagem_musicas.sort_values(by='contagem_title', ascending=False)
df_contagem_musicas
```
