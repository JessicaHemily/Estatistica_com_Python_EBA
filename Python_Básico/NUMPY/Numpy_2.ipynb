```python
import numpy as np
```

Criar vetor:

```python
a = np.arange(6)
print(a)
```

Gerar um vetor 3x2, partindo do vetor a:

```python
b = a.reshape(3,2)
print(b)
```

Concatenar vetores:

```python
vet1 = np.array([1,3,5,7])
vet2 = np.array([2,4,6,8])

vet3 = np.concatenate((vet1,vet2))
print(vet3)
```

Ordenar o vetor vet3:

```python
vet_ordenado = np.sort(vet3)
print(vet_ordenado)
```
